```python
# 4. Running  FastQC
!mkdir -p /content/fastqc_pre
!fastqc SRR531365_1.fastq.gz SRR531365_2.fastq.gz -o /content/fastqc_pre
!ls -lh /content/fastqc_pre

from google.colab import files
files.download("/content/fastqc_pre/SRR531365_1_fastqc.html")
files.download("/content/fastqc_pre/SRR531365_2_fastqc.html")
```

```python
#  5. Preparing the  Trimmomatic Adapter File
!rm -rf /usr/local/bin/trimmomatic*
!rm -rf trimmomatic-0.39.zip Trimmomatic-0.39
!wget -q https://github.com/usadellab/Trimmomatic/releases/download/v0.39/trimmomatic-0.39.zip
!unzip -q trimmomatic-0.39.zip
!mv Trimmomatic-0.39 /usr/local/bin/
!echo 'java -jar /usr/local/bin/Trimmomatic-0.39/trimmomatic-0.39.jar "$@"' > /usr/local/bin/trimmomatic
!chmod +x /usr/local/bin/trimmomatic

!mkdir -p /usr/local/bin/Trimmomatic-0.39/adapters
!wget -q https://raw.githubusercontent.com/timflutre/trimmomatic/master/adapters/TruSeq3-PE.fa -O /usr/local/bin/Trimmomatic-0.39/adapters/TruSeq3-PE.fa
```

```python
# 6. Trimming the Reads with Trimmomatic tool
!mkdir -p /content/trimmed
!trimmomatic PE -threads 2 \
SRR531365_1.fastq.gz SRR531365_2.fastq.gz \
/content/trimmed/SRR531365_1_paired.fq.gz /content/trimmed/SRR531365_1_unpaired.fq.gz \
/content/trimmed/SRR531365_2_paired.fq.gz /content/trimmed/SRR531365_2_unpaired.fq.gz \
ILLUMINACLIP:/usr/local/bin/Trimmomatic-0.39/adapters/TruSeq3-PE.fa:2:30:10 \
LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36
```

```python
# 7. Trimming the  Reads with Cutadapt tool
!mkdir -p /content/cutadapt_out
!cutadapt -a AGATCGGAAGAGC -A AGATCGGAAGAGC \
-o /content/cutadapt_out/SRR531365_1_trimmed.fq.gz \
-p /content/cutadapt_out/SRR531365_2_trimmed.fq.gz \
SRR531365_1.fastq.gz SRR531365_2.fastq.gz
```

Here is an interpretation of the Cutadapt output from the previous cell:

*   **`Total read pairs processed: 3,154,678`**: This is the total number of paired-end reads that were analyzed by Cutadapt.
*   **`Read 1 with adapter: 125,936 (4.0%)`**: This indicates that 125,936 reads in the first file of the paired-end data contained the specified adapter sequence, representing 4.0% of the total read 1 pairs.
*   **`Read 2 with adapter: 101,301 (3.2%)`**: This indicates that 101,301 reads in the second file of the paired-end data contained the specified adapter sequence, representing 3.2% of the total read 2 pairs.
*   **`Pairs written (passing filters): 3,154,678 (100.0%)`**: This shows that all 3,154,678 read pairs passed the filtering criteria (which in this case, with the provided parameters, primarily means the adapter trimming was performed, and no reads were discarded based on length or other quality metrics not specified in the command).
*   **`Total basepairs processed: 637,244,956 bp`**: This is the total number of bases across all read pairs before trimming.
*   **`Total written (filtered): 631,397,995 bp (99.1%)`**: This is the total number of bases remaining after trimming, which is 99.1% of the original base pairs. This indicates that a small percentage of bases were removed due to adapter trimming.
*   **`Read 1: 314,515,227 bp`**: The number of base pairs remaining in Read 1 after trimming.
*   **`Read 2: 316,882,768 bp`**: The number of base pairs remaining in Read 2 after trimming.
*   The subsequent sections ("=== First read: Adapter 1 ===" and "=== Second read: Adapter 2 ===") provide a detailed breakdown of the adapter trimming for each read in the pair, including the sequence of the adapter, the number of times it was trimmed, the minimum overlap required, the number of allowed errors during matching, the nucleotide composition preceding the removed adapters, and a detailed table showing the length distribution of the removed adapter sequences and the number of errors found for each length. This detailed breakdown helps in understanding the efficiency and accuracy of the adapter trimming process.

Here is a comparison of the trimming results from Trimmomatic and Cutadapt based on the provided outputs:

**Trimmomatic Output Summary:**

*   **Input Read Pairs:** 3,154,678
*   **Both Surviving:** 482,081 (15.28%) - These are the read pairs where both reads survived the trimming process and were kept as a pair.
*   **Forward Only Surviving:** 137,077 (4.35%) - These are the read pairs where only the forward read survived trimming.
*   **Reverse Only Surviving:** 15,159 (0.48%) - These are the read pairs where only the reverse read survived trimming.
*   **Dropped:** 2,520,361 (79.89%) - These are the read pairs that were completely discarded by Trimmomatic.

**Cutadapt Output Summary:**

*   **Total read pairs processed:** 3,154,678
*   **Read 1 with adapter:** 125,936 (4.0%) - Number of Read 1 reads where an adapter was found and trimmed.
*   **Read 2 with adapter:** 101,301 (3.2%) - Number of Read 2 reads where an adapter was found and trimmed.
*   **Pairs written (passing filters):** 3,154,678 (100.0%) - All read pairs were written to output files after trimming.
*   **Total basepairs processed:** 637,244,956 bp
*   **Total written (filtered):** 631,397,995 bp (99.1%) - Total base pairs remaining after trimming.

**Comparison:**

*   **Read Retention:** The most significant difference is in the number of read pairs retained. Trimmomatic discarded a large percentage of read pairs (79.89%), while Cutadapt retained all read pairs (100.0%). This suggests that Trimmomatic's filtering criteria (LEADING:3, TRAILING:3, SLIDINGWINDOW:4:15, MINLEN:36) were more stringent, leading to more reads being dropped. Cutadapt, with the parameters used, primarily performed adapter trimming and did not discard reads based on quality or length as Trimmomatic did.
*   **Adapter Trimming:** Both tools identified and trimmed adapter sequences. Cutadapt reported that adapters were found in 4.0% of Read 1 and 3.2% of Read 2. Trimmomatic's output doesn't explicitly state the percentage of reads with adapters, but the "Both Surviving", "Forward Only Surviving", and "Reverse Only Surviving" counts, along with the "Dropped" count, are a result of adapter trimming *and* other quality filtering steps.
*   **Output Files:** Trimmomatic produced four output files for paired-end data (paired forward, unpaired forward, paired reverse, unpaired reverse), reflecting its ability to handle reads where only one of the pair survives trimming. Cutadapt, with the parameters used, produced two output files for paired-end data (trimmed forward and trimmed reverse), indicating that it kept all pairs together after trimming.
*   **Focus:** Trimmomatic is a more comprehensive trimmer that can perform various quality trimming steps in addition to adapter removal. Cutadapt is primarily focused on adapter trimming and filtering based on length and quality, but the command used here only specified adapter trimming.

In summary, Trimmomatic applied more aggressive quality and length filtering, resulting in a higher number of dropped reads and the separation of surviving paired and unpaired reads. Cutadapt, with the parameters used, focused solely on adapter trimming and retained all read pairs, only reducing the total number of base pairs due to the removal of adapter sequences. The choice of which tool and parameters to use depends on the specific goals of the sequencing analysis and the desired level of quality control.
